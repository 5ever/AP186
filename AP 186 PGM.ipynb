###Filled Circle

```python
import numpy as np

n = 30
m = 30

circle = np.zeros(shape = ((n+1),m))
xgrid, ygrid = 15,15
r = 1.2

epsilon = 5

#map_ = [['.' for x in range(n)] for y in range(m)]

for y in range(m):
    for x in range(n):
        if abs ((x-xgrid)**2 + (y-ygrid)**2 - r**2) < epsilon**2:
            circle[y][x] = 1
            
pgmeheader = 'pgm' + '\n' + str(n) + ' ' + str(m) + '\n' + '255' + '\n' + '\n'
f = open('circle.txt', 'w+')
f.write(pgmeheader)
np.savetxt('circle.txt', a, fmt='%d');
f.close()


fig = plt.figure
plt.imshow(circle, cmap = 'gray')
plt.show()


```

###Concetric Circles with different gray levels

```python
import numpy as np

n = 30
m = 30

concentric = np.ones(shape = ((n+1),m))
xgrid, ygrid = 15,15
r = 4

epsilon = 3

for y in range(m):
    for x in range(n):
        if abs ((x-xgrid)**2 + (y-ygrid)**2 - r**2) < epsilon**2:
            concentric[y][x] = 0            
#concentric = cicle
pgmeheader = 'pgm' + '\n' + str(n) + ' ' + str(m) + '\n' + '255' + '\n' + '\n'
f = open('concentric.txt', 'w+')
f.write(pgmeheader)
np.savetxt('concentric.txt', concentric, fmt='%d');
f.close()


fig = plt.figure
plt.imshow(concentric, cmap = 'gray')
plt.show()

```

###Load the plain PGM file of your image using your reading function and display its properties, such as image dimensions and number of gray levels. Most images we encounter every day are saved in 24-bit color which easily reduces to 8-bit grayscale, meaning 8 bits are associated with a pixel’s value, allowing one to record up to 256 distinct gray level values.

```python
im = cv2.imread('2dPGM.PGM')
height, width = im.shape[:2]
hist = cv2.calcHist([im],[0],None,[256],[0,256])
print "height " + str(height) +'\n',"width "+ str(width) + '\n', "data type "+str(im.dtype)
x = range(256)
plt.bar(x,hist)
plt.xlim(0,255)
plt.show()
```
